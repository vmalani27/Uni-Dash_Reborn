```python
import pandas as pd
```

```python
df_path = "C:\\Users\\Vansh Malani\\Documents\\GitHub\\Uni-Dash_Reborn\\data"
df = pd.read_csv(df_path + "\\23DCS023_all_emails.csv")
```

```python
df.head()
```

```python
import re

def clean_text(text):
    # Remove disclaimers, greetings, signatures, and repeated line breaks
    text = re.sub(r"(?i)(dear student|regards|thank you|disclaimer:.*|sent from my.*|best wishes|cheers|sincerely|yours truly|confidentiality notice:.*)", "", text)
    text = re.sub(r"\n{2,}", "\n", text)  # Replace multiple line breaks with one
    text = re.sub(r"\s{2,}", " ", text)  # Replace multiple spaces with one
    return text.strip()

# Drop useless columns
cols_to_drop = ["Unnamed: 0", "id", "threadId", "labelIds", "snippet"]
df = df.drop(columns=cols_to_drop)

# Combine subject + content
if "subject" in df.columns and "content" in df.columns:
    df["full_text"] = df["subject"].astype(str) + " " + df["content"].astype(str)
else:
    raise ValueError("Required columns missing: subject and/or content")

# Clean the combined text
if "full_text" in df.columns:
    df["full_text"] = df["full_text"].apply(clean_text)
else:
    raise ValueError("full_text column missing after combination")

# Keep only what matters
df = df[["from", "full_text"]]
df["label"] = ""  # Placeholder for manual labeling
df.head()
```
