```python
# --------------------------------------------------------
# Regex-based academic semantic detection (NO substrings)
# Updated: Coursework, syllabus, timetable, etc. ‚Üí Lecture
# --------------------------------------------------------

ASSIGNMENT_PATTERNS = [
    r"\bassignment\b",
    r"\bassignments\b",
    r"\bhomework\b",
    r"\bsubmit\b",
    r"\bsubmission\b",
    r"\bsubmit by\b",
    r"\bdue\b",
    r"\bdue by\b",
    r"\bdeadline\b",
    r"\bturn in\b",
    r"\bupload\b"
]

TEST_PATTERNS = [
    r"\bquiz\b",
    r"\bquizzes\b",
    r"\bviva\b",
    r"\bunit test\b",
    r"\bclass test\b",
    r"\binternal test\b",
]

EXAM_PATTERNS = [
    r"\bexam\b",
    r"\bexamination\b",
    r"\bmid[- ]?sem\b",
    r"\bend[- ]?sem\b",
    r"\bms/es\b",
    r"\bseating plan\b",
    r"\bhall ticket\b",
    r"\bexam timetable\b",
    r"\btimetable\b",
    r"\bexam schedule\b",
    r"\bschedule\b",
]

# Lecture: includes coursework, syllabus, timetable, class/lab info, etc.
LECTURE_PATTERNS = [
    r"\bclass\b",
    r"\blecture\b",
    r"\brescheduled\b",
    r"\bcancelled\b",
    r"\bpresentation session\b",
    r"\bpractical session\b",
    r"\btheory class\b",
    r"\blab session\b",
    r"\bsyllabus\b",
    r"\bunit plan\b",
    r"\bweekly schedule\b",
    r"\bteaching plan\b",
    r"\bpractical list\b",
    r"\broom changed\b",
    r"\bclass updates?\b",
    r"\bclass cancellation\b",
    r"\bfaculty announcement\b",
    r"\bsemester plan\b",
    r"\blecture notes?\b",
    r"\bclass schedule\b",
    r"\blab schedule\b",
    r"\bunit test schedule\b",
    r"\bsubject plan\b",
    r"\bsubject outline\b"
]

URGENT_PATTERNS = [
    r"\burgent\b",
    r"\bimmediate action\b",
    r"\blast date\b",
    r"\btoday only\b",
    r"\bimportant update\b",
]

PROGRAM_PATTERNS = [
    r"\bacademy\b",
    r"\bbatch\b",
    r"\bprogram\b",
    r"\bcertificate\b",
    r"\bcertification\b",
    r"\benroll\b",
    r"\benrollment\b",
    r"\bjoining\b",
    r"\bworkshop series\b",
    r"\bseries\b"
]

NEWSLETTER_PATTERNS = [
    r"\bunsubscribe\b",
    r"\bnewsletter\b",
    r"\bpromo\b",
    r"\bpricing\b",
    r"\boffer\b",
    r"\bsale\b",
    r"\bbilling\b",
    r"\bdiscount\b",
]

EVENT_PATTERNS = [
    r"\bctf\b",
    r"\bcapture the flag\b",
    r"\bhackathon\b",
    r"\bworkshop\b",
    r"\bwebinar\b",
    r"\bseminar\b",
    r"\bcompetition\b",
    r"\bchallenge\b",
    r"\bguest lecture\b",
    r"\btech talk\b",
    r"\bboot ?camp\b",
    r"\bhands[- ]on\b",
    r"\btraining\b",
    r"\bcoding contest\b",
    r"\bhack ?day\b",
]

def regex_match(text: str, patterns) -> bool:
    if not isinstance(text, str):
        return False
    t = text.lower()
    return any(re.search(p, t) for p in patterns)

def is_assignment(text): return regex_match(text, ASSIGNMENT_PATTERNS)
def is_test(text): return regex_match(text, TEST_PATTERNS)
def is_exam(text): return regex_match(text, EXAM_PATTERNS)
def is_lecture(text): return regex_match(text, LECTURE_PATTERNS)
def is_urgent(text): return regex_match(text, URGENT_PATTERNS)
def is_program(text): return regex_match(text, PROGRAM_PATTERNS)
def is_newsletter(text): return regex_match(text, NEWSLETTER_PATTERNS)
def is_event(text): return regex_match(text, EVENT_PATTERNS)

def is_academic_keyword(text):
    return (
        is_assignment(text) or
        is_test(text) or
        is_exam(text) or
        is_lecture(text)
    )
```
